#### Plot bar graph between alcohol and quality

```python
plt.figure(figsize=[4,2])
# plot bar graph
plt.bar(data['quality'],data['alcohol'],color='red')
# label x-axis
plt.xlabel('quality')
#label y-axis
plt.ylabel('alcohol')
plt.show()
```

The above bar graph helps to find which quality wine has highes alcohol.
